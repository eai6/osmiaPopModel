```python
############## check dates that are not downloaded and download the data ###########

data_folder = '/Users/edwardamoah/Documents/GitHub/OsmiaPopModel/data/prism/'
```

```python
import os
```

```python
files = os.listdir(data_folder)
```

```python
data_type = []
date = []
for file in files:
    try:
        if len(file.split('_')) > 3:
            data_type.append(file.split('_')[1])
            date.append(file.split('_')[4])
    except IndexError:
        print("Error: Invalid file format")
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'data_type':data_type, 'date':date})
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
```

```python
import pandas as pd

start_date = '2008-01-01'
end_date = '2021-12-31'

dates = pd.date_range(start=start_date, end=end_date, freq='D')
```

```python
ppt_dates = df[df['data_type'] == 'ppt']['date'].tolist()
tmean_dates = df[df['data_type'] == 'tmean']['date'].tolist()
```

```python
tmean_no_download = []
ppt_no_download = []
for date in dates:
    if date not in ppt_dates:
        ppt_no_download.append(date)
        #print(date)
    if date not in tmean_dates:
        tmean_no_download.append(date)
        #print(date)
```
